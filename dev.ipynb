```python
import matplotlib.pyplot as plt
```

```python
a = [0.05, 0.1, 0.2, 0.4, 0.5]
time = [394, 398, 413, 421, 430]
acc = [0.371, 0.519, 0.455, 0.395, 0.401]
```

```python
fig, ax = plt.subplots(figsize=(7, 5))  # Slightly larger figure

plt.plot(a, acc, "bo-", linewidth=2, markersize=6, label="Accuracy")  # Blue line with markers
ax.set_xlabel(r"$a$", fontsize=14)  # Larger x-axis label
ax.set_ylabel("Accuracy (%)", fontsize=14)  # Larger y-axis label
ax.tick_params(axis='both', which='major', labelsize=12)  # Larger tick labels
ax.grid(True, linestyle="--", alpha=0.6)  # Light grid for readability
plt.legend(fontsize=12)  # Add legend
plt.savefig("images/acc.png", dpi=600, bbox_inches="tight")  # High DPI and tight layout
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))

plt.plot(a, time, "ko--")
ax.set_xlabel(r"$a$")
ax.set_ylabel("Time (s)")
plt.savefig("images/time.png", dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))  # Increase figure size for better readability

plt.plot(a, time, "bo-", linewidth=2, markersize=6, label="Computation Time")  # Blue solid line with larger markers
ax.set_xlabel(r"$a$", fontsize=14)  # Increase font size
ax.set_ylabel("Time (s)", fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=12)  # Increase tick label size
ax.grid(True, linestyle="--", alpha=0.6)  # Add light grid for readability
plt.legend(fontsize=12)  # Add legend
plt.savefig("images/time.png", dpi=600, bbox_inches="tight")  # Higher DPI and tight layout
plt.show()
```
